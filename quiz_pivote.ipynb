```python
path='/content/drive/MyDrive/5to semestre/1er periodo/Visualización de datos para la toma de decisiones/Pivot Tables/antropometria-dataset-2.csv'
import pandas as pd
import numpy as np

df=pd.read_csv(path)
df.head()
```

```python
df.columns
```

```python
 countsDF = pd.pivot_table(data=df,
               index='desc_ent',
               values='peso',
               #aggfunc='count'
               )
```

```python
countsDF
```

```python
 counts = pd.pivot_table(data=df,
               index=['desc_ent','sexo'],
               values='peso',
               #aggfunc='count'
               )
 counts
```

```python
 countsD = pd.pivot_table(data=df,
               index=['desc_ent','sexo'],
               values='peso',
               aggfunc='median'
               )
 countsD
```
