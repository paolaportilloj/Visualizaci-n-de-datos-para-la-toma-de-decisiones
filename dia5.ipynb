```python
p='https://raw.githubusercontent.com/tec03/Datasets/main/datasets/satis.csv'
import pandas as pd
import numpy as np

df=pd.read_csv(p)
df.head()
```

```python
df.isnull().any().sum
```

```python
df.shape
```

```python
df.dropna(how='all',inplace=True)
```

```python
df.shape
```

```python
df.tail()
```

```python
df.head(2)
```

```python
def custom_cls(x):
  if 0<=x<=20:
    return '<=20'
  elif 21<=x<=30:
    return '21-30'
  elif 31<=x<=40:
    return '31-40'
  return '>40'

```

```python
custom_cls(-5)
```

```python
df['AgeGP']=df.apply(lambda tbl:custom_cls(tbl['Age']),
                                           axis=1)
df.head()
```

```python
 countsDF = pd.pivot_table(data=df,
               index='AgeGP',
               values='Price',
               aggfunc='count'
               )
```

```python
countsDF=countsDF.rename({'Price':'Counts'},
                     axis=1)
countsDF
                     
```

```python
countsDF.columns
```

```python
countsDF.reset_index(inplace=True)
```

```python
t=countsDF.loc[:,'Counts'].sum()
```

```python
countsDF['%']=countsDF.loc[:,'Counts']
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

```

```python
cols=['gold', 'skyblue', 'yellowgreen', 'coral', 'pink']
explode_dis=[0.04,0.04,0.06,0.03]
countsDF['Counts'].plot(kind='pie',
                        figsize=(15,6),
                        startangle=90,
                        labels=None,
                        autopct='%1.1f%%',
                        pctdistance=1.1,
                        colors=cols,
                        explode=explode_dis,
                        shadow=True,
)
plt.title('Number of participants, by age Group')
plt.legend(labels=countsDF.AgeGP,
           bbox_to_anchor=(1,1)
           );
```
