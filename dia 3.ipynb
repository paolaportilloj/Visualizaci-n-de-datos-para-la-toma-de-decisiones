```python
import pandas as pd
import numpy as np

data= {'Company':['Ford', 'Ford', 'VW','Cooper','BMW','Cooper'],
       'Stars':[1,2,3, np.nan, 1,2 ],
       'Weight': [2,4,3,5,2, None ],
       'Origin': [ 'China', 'Mexico', 'China', None, 'Mexico' , np.nan],
       'Length':[ 40, 50, 30, np.nan, 45,pd.NaT]
       }
```

```python
df= pd.DataFrame(data)
df
```

```python
df.isnull().any()
```

```python
df.isnull().any(axis=1)
```

```python
ndf=df.copy()
ndf
ndf.dropna(inplace=True)
ndf
```

```python
ndf2=df.copy()
ndf2
```

```python
ndf2.dropna(how='all',inplace=True)
ndf2
```

```python
ndf2.dropna(thresh=3, inplace=True)
ndf2
```

```python
df.isnull().any(axis=1)
```

```python
favs=df.isnull().any(axis=1)
```

```python
df[ favs]
```

```python
sm=df.Stars.mean()
sm
df[ 'Stars']=df['Stars' ].fillna(value=sm)
df
```

```python
wm=df.Weight.mean()
df[ 'Weight']=df['Weight' ].fillna(value=wm)
df
```

```python
om=df.Origin.mode()[ ]
```

```python
df[ 'Origin']=df[ 'Origin'].fillna(value=om)
```

```python
df
```

```python
my_pivot=pd.pivot_table(df,
               index= ['Origin'],
               values= 'Length',
               aggfunc= [ 'sum','median','mean'],
               margins=True,
               columns='Company'
              
               )
```
